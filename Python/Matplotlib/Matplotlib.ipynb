```python
y=np.sin(x)
```

```python
fig,axes=plt.subplots()
axes.plot(x,y,"b")
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,"b")
axes[1].plot(y,x,"r")
```

```python
fig,(axes1,axes2)=plt.subplots(nrows=1,ncols=2)
axes1.plot(x,y,"b")
axes2.plot(y,x,"r")
```

```python
fig,axes=plt.subplots(2,4)
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x))
axes[0,2].plot(x,np.tan(x),"y")
axes[0,3].plot(x,np.cos(x)**2,"r")

```

```python
fig,((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8))=plt.subplots(2,4)
ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x))
ax3.plot(x,np.tan(x),"y")
ax4.plot(x,np.cos(x)**2,"r")
fig.tight_layout()
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,5))
axes[0].plot(x,y,label="$sin(x)$")
axes[0].set_title("Relacin X - Y")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].legend(loc="lower left")
axes[1].plot(y,x)
axes[1].set_title("Relacin Y - X")
axes[1].set_xlabel("Y")
axes[1].set_ylabel("X")
```

```python
#matplot
fig,ax=plt.subplots(figsize=(8,8))
ax.plot(x,x+1,"r-")
ax.plot(x,x+2,"b-x")
ax.plot(x,x+3,"y:")
ax.plot(x,x+4,"g-.")
```

```python
#pyplot
fig,ax=plt.subplots(figsize=(8,8))
ax.plot(x,x+1,color="red", alpha=0.1, linewidth=5, linestyle="-")
ax.plot(x,x+2,color="blue", linestyle="--")
ax.plot(x,x+3,color="#66FF66", linestyle="dashed") #RGB CODE
ax.plot(x,x+4,color="black", linestyle=":", marker="o", markersize=8, markerfacecolor="#CCFFCC")
```

Other graphs

```python
data=np.random.randint(1,50,100)
```

```python
#histogram
plt.hist(data,bins=5,rwidth=0.5,histtype="step")
plt.show()
```

```python
#boxplot
plt.boxplot(data,vert=False,patch_artist=True,notch=True)
plt.show()
```

```python
data=np.append(data,200)
plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=True)
plt.show()
```

```python

plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=False)
plt.show()
```
