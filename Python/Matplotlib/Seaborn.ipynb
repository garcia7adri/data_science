```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="darkgrid")
sns.barplot(x=["A","B","C"],y=[1,2,3])
plt.show()
```

```python
sns.set(style="dark",palette="muted")
sns.barplot(x=["A","B","C"],y=[1,2,3])
plt.show()
```

```python
sns.set(style="dark",palette="dark")
sns.barplot(x=["A","B","C"],y=[1,2,3])
plt.show()
```

```python
sns.set(style="dark",palette="muted",font="Verdana",font_scale=1)
sns.barplot(x=["A","B","C"],y=[1,2,3])
plt.show()
```

```python
#preloaded data
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.displot(data=tips,x="total_bill")
plt.show()
```

```python
sns.displot(data=tips,x="total_bill",y="tip")
plt.show()
```

```python
sns.displot(data=tips,x="total_bill",y="tip",hue="sex")
plt.show()
```

```python
sns.displot(data=tips,x="total_bill",hue="sex")
plt.show()
```

```python
sns.displot(data=tips,x="total_bill",hue="sex",kind="kde")
plt.show()
```

```python
sns.displot(data=tips,x="total_bill",hue="sex",kind="kde",legend=True,palette="dark")
plt.show()
```

```python
sns.displot(data=tips,x="total_bill",hue="sex",kind="kde",legend=True,palette="dark",alpha=.25)
plt.show()
```

```python
sns.histplot(data=tips,x="tip",bins=15, cumulative=False,hue="sex",stat="count",multiple="dodge")
plt.show()
```

```python
sns.kdeplot(tips,x="tip",hue="sex", cumulative=False, fill=True, bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(tips,x="tip",hue="sex",stat="count")
plt.show()
```

```python
sns.displot(tips,x="tip",hue="sex",kind="hist",multiple="stack")
plt.show()
```

```python
sns.countplot(data=tips,x="day",hue="sex")
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips,x="day",y="total_bill",hue="sex",dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips,x="day",y="total_bill",hue="sex",dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips,x="day",y="total_bill",hue="sex",dodge=True)
sns.swarmplot(data=tips,x="day",y="total_bill",hue="sex",dodge=True,palette="dark:0",marker="<")
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips,x="day",y="total_bill",hue="sex",split="True",
               dodge=True,palette="pastel")

plt.show()
```

```python
sns.catplot(data=tips,x="day",y="total_bill",hue="sex",dodge=True,kind="swarm")
plt.show()
```

```python
sns.catplot(data=tips,x="day",y="total_bill",hue="sex",dodge=True,kind="box")
plt.show()
```

```python
sns.catplot(data=tips,x="day",y="total_bill",hue="sex",dodge=True,kind="box",col="time")
plt.show()
```

```python
plt.figure(figsize=(8,8))
markers={"Lunch":"D","Dinner":"s"}
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",style="time",size="size")
plt.legend(loc="center",bbox_to_anchor=(1.12,0.6))
plt.show()
```

```python
sns.lineplot(data=tips,x="total_bill",y="tip",hue="day",style="time",size="size")
plt.show()
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",hue="day",style="time",size="size",col="time")
plt.show()
```

```python
sns.jointplot(data=tips,x="total_bill",y="tip",hue="sex",kind="hist")
plt.show()
```

```python
sns.jointplot(data=tips,x="total_bill",y="tip",hue="sex",kind="kde")
plt.show()
```

```python
sns.jointplot(data=tips,x="total_bill",y="tip",hue="sex",kind="hist",marginal_ticks=True, 
              marginal_kws=dict(bins=25, fill=False,multiple="dodge"))
plt.show()
```

```python
sns.pairplot(data=tips,hue="sex",palette="coolwarm",corner=True,kind="scatter")
plt.show()
```
