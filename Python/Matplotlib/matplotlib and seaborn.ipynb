```python
#Juxtaposing two different data for comparison on a bar-plot
name=["John","Sally","Harry","Smith"]
tv_watch,sns_use=([40,50,60,70],[70,60,50,40])
name_i=np.arange(len(name))
width=0.2

plt.figure(figsize=(10,5))
plt.bar(name_i-width/2,tv_watch,width, color="r",label="TV")
plt.bar(name_i+width/2,sns_use,width, color="g",label="SNS")
plt.xticks(name_i,name)
plt.title("Time spent by hours in media")
plt.ylabel("Minutes/week")
plt.legend(loc="upper center")
plt.grid()
plt.show()
```

```python
x=np.random.normal(-7,3,1000) #(mean, std, random values for distribution)
y=np.random.normal(0,3,1000)
z=np.random.normal(7,3,1000)
bins=np.linspace(-20,20,50) #(min,max,number of bins)

fig,ax=plt.subplots()
ax.hist([x,y,z],bins,label=["x","y","z"],color=["r","g","b"])
ax.legend(loc="upper right")
plt.show()
```

```python
def stnd_func(mu,sigma,x):
    """Returns standard normal distribution probability desnsity at x. """
    
    return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-(x-mu)**2/(2*sigma**2))
```

```python
mu=100;sigma=10;sample=1000
pdf=np.random.normal(mu,sigma,sample)
bins=np.linspace(60,140,80)


fig,ax=plt.subplots()
ax.hist(pdf,bins,color="g")
ax.plot(bins,stnd_func(mu,sigma,bins)*sample,"r")
plt.show()
```

```python
setosa=iris.iloc[0:50,:]
versicolor=iris.iloc[50:100,:]
virginica=iris.iloc[100:150,:]
names=["Setosa","Versicolor","Virginica"]
kinds=[setosa,versicolor,virginica]
```

```python
fig,ax=plt.subplots()
ax.hist([setosa.SepalL,versicolor.SepalL,virginica.SepalL],bins=20,label=names,color=["r","g","b"])
ax.set_title("Distribution of sepal by kinds")
ax.set_xlabel("Sepal Length(cm)",fontsize=10)
ax.set_ylabel("Frequency",fontsize=10)
ax.legend(fontsize=10,loc="upper right")
plt.show()
```

Scatterplot, pie chart and boxplot

```python
#scatterplot
fig,ax=plt.subplots(figsize=(8,5))
mk_size=40
ax.scatter(setosa.PetalL,setosa.PetalW,marker="*",s=mk_size,c="r",edgecolor="black",label="Setosa")
ax.scatter(versicolor.PetalL,versicolor.PetalW,marker="+",s=mk_size,c="g",label="Versicolor")
ax.scatter(virginica.PetalL,virginica.PetalW,marker=">",s=mk_size,c="b",label="Virginica")

ax.set_xlabel("Petal Length(cm)",fontsize=12)
ax.set_ylabel("Petal Widht(cm)",fontsize=12)
ax.legend(loc="upper left", fontsize=14)
plt.show()
```

```python
#Using for-loop in plotting multiple plots
fig,ax=plt.subplots(figsize=(6,5))
for name, kind in zip(names, kinds):
    ax.plot(kind.PetalL,kind.PetalW,marker="*",ms=10,linestyle="",markeredgecolor="black",label=name)
ax.legend(loc="upper left", fontsize=13)
ax.set_title("Distribution of Iris by its kinds",fontsize=14)
ax.set(xlabel="Petal Length(cm)",ylabel="Petal Widthc(cm)")
plt.show()
```

```python
#boxplot
fig,ax=plt.subplots(figsize=(8,5))
ax.boxplot(np.array(iris.iloc[:,0:4]),sym="b^",showmeans=True) #np.array if dataframe is not working
ax.set_xticklabels(iris.columns[:4])
ax.set_title("Iris data box plot")
ax.set_ylabel("cm")
plt.show()
```

```python
#pie chart
name=["John","Sally","Harry","Smith"]
tv_watch,sns_use=([30,40,60,70],[70,60,40,30])
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(8,4))
ax1.pie(tv_watch,explode=(0,0.1,0,0),labels=name,textprops={"fontsize":14}, autopct="%.2f%%",shadow=True,startangle=90,
        colors=["r","g","c","orange"])
ax2.pie(sns_use,labels=name,autopct="%.1f%%",shadow=True,startangle=180,wedgeprops={"width":0.6,"edgecolor":"w","linewidth":2})
ax1.set_title("Tv watching behavior",fontsize=14)
ax2.set_title("SNS use behaviour",fontsize=14)
ax1.axis("equal")
fig.tight_layout()
plt.show()
```
