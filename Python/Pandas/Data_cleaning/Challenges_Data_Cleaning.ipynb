```python
import pandas as pd   #Tratamiento de datos
import numpy as np    #Operar numericamente sobre datos

import datetime       #Manejo y tratamiento de Fechas

```

```python
movies = pd.read_csv('https://raw.githubusercontent.com/Juan199963/Python_Course/main/movie_metadata.csv')
```

## Si desplegamos la columna title_year encontraremos que efectivamente se trata como un string sin punto decimal al final y en los campos donde no hay valores aparece NaN, reemplacemos lo valores NaN

```python
movies['title_year']
#movies.dtypes
```

```python
movies['title_year']=pd.to_numeric(movies['title_year'], errors='coerce')#.astype('int64')
```

```python
movies['title_year'] = movies['title_year'].fillna(value=0)
movies['title_year'].astype('int64')
```

## Reemplaza la duration del filme con la media



```python
movies['duration']= movies['duration'].fillna(movies['duration'].median())
```

## Si no es necesario conservar los decimales de la media, convierte esta columna en tipo de dato entero

```python
movies['duration'].astype('int64')
```
