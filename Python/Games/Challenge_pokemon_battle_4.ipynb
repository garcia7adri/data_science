```python

class Movements:
            """Clase de los movimientos"""
            def __init__(self, nombre, poder, tipo):
                self.nombre = nombre
                self.poder = poder
                self.tipo = tipo


class Pokemon:
    """Creamos la clase pokemon, con la cual podemos registrar a los pokemons que utilizaremos para \
    simular un combate"""
    def __init__(self, especie, tipo, ataque, defensa, vida, velocidad, fortalezas, debilidades, movimientos):
        self.especie = especie
        self.tipo = tipo
        self.ataque = ataque
        self.defensa = defensa
        self.vida = vida
        self.vida_orig = vida
        self.velocidad = velocidad
        self.fortalezas = fortalezas
        self.debilidades = debilidades
        self.movimientos = movimientos

   

    
    def centro_pokemon(self):
        """La funcion centro_pokemon nos permite restablecer el valor de vida""" 
        self.vida = self.vida_orig
        
    
    def pelea(self, rival):
        """La funcion pelea nos permite elegir 2 pokemon para iniciar un combate.\ 
        Inicia quien tenga la velocidad mayor, en caso de empate va el primer pokemon"""
        mi_turno = self.velocidad >= rival.velocidad

         
        round=1
        while (self.vida > 0) & (rival.vida > 0):
            """Utilizando un ciclo while le indicamos a la funcion que siga corrinedo hasta \
        que la vida de alguno de los pokemon llegue a 0"""
            
            """Con estos print hacemos visibles el numero de ronda en el que nos encontramos \
            y la vida de los pokemon al momento de iniciar la ronda"""  
            
            print(f"\nROUND # {round}\n")
            print(f"Vida de {self.especie.title()} = {self.vida} \n") 
            print(f"Vida de {rival.especie.title()} = {rival.vida}\n" )
            
            if mi_turno:
                """Con el siguiente ciclo while pedimos que se seleccione uno de los \
                movimientos que tiene el pokemon, y en caso de que se seleccione una opcion \
                no valida el programa indicara que no es una opcion valida y pedira que se \
                seleccione una opcion nuevamente hasta que se seleccione una correcta"""
                while True:
                
                    print(f"------------{self.especie.upper()} ATACA------------\n")
                    print(f"¿Qué tipo de ataque quieres? \n")

                    for movimiento in self.movimientos:
                        print(f"{movimiento.title()}")
                    tipo_ataque = input(f"\n")
                    tipo_ataque=tipo_ataque.lower().strip()

                    if tipo_ataque in self.movimientos:
                        print(f"\nHas elegido {tipo_ataque}.\n")
                        break
                    else:
                        print(f"\nNo es una opción valida. Escribe un movimiento disponible.\n")
                                      
                """Despues de seleccionar un movimiento calculamos el daño que le hacemos \
                al rival, por lo que formulamos una ecuacion, el poder del moviento por el \
                cociente del ataque de nuestro pokemon y la defensa del rival""" 
                
                #print(self.movimientos[tipo_ataque].poder)
                mi_dano = (self.movimientos[tipo_ataque].poder)* (self.ataque / rival.defensa)
                print(f"\nEl daño que puede causar es {mi_dano} \n")
                
                """Una vez calculado el daño, se tiene que multiplicar por el modificador \
                que corresponda dependiendo si el tipo del movimiento se encuentra entre \
                las debilidades del pokemon rival, y una vez calculado todo, lo restampos \
                de la vida del oponente"""
                if self.movimientos[tipo_ataque].tipo in rival.debilidades:
                    mi_mult = 2
                elif self.movimientos[tipo_ataque].tipo in rival.fortalezas:
                    mi_mult = .5
                else:
                    mi_mult = 1
                print(f"El real que causará según las debilidades y fortalezas es {int(mi_dano * mi_mult)} \n")
                rival.vida = rival.vida - int(mi_dano * mi_mult)
              
                print(f"{self.especie} ha hecho {int(mi_dano * mi_mult)} puntos de daño \n")
                print(f"A {rival.especie} le quedan {rival.vida} puntos de vida\n")
                
                """Una vez calculada la vida restante del pokemon se da por terminado el turno"""
                mi_turno = not mi_turno          
                        
            else:
                """A partir de aquí el turno del rival comienza, y 
                consiste basicamente del mismo codigo usado anteriormente"""
                
                while True:
                    
                    print(f"------------{rival.especie.upper()} ATACA ------------\n")
                    print(f"¿Qué tipo de ataque quieres? \n")
                    for movimiento in rival.movimientos:
                        print(f"{movimiento.title()}\n")
                    tipo_ataque = input(f"\n")
                    tipo_ataque=tipo_ataque.lower().strip()

                    if tipo_ataque in rival.movimientos:
                        print(f"\nHas elegido {tipo_ataque}.\n")
                        break
                    else:
                        print(f"\nNo es una opción valida. Escribe un movimiento disponible.\n")
               
                #print(rival.movimientos[tipo_ataque].poder)
                rival_dano = (rival.movimientos[tipo_ataque].poder)*(rival.ataque / self.defensa)
                print(f"\nEl daño que puede causar es {rival_dano} \n")
                
                if rival.movimientos[tipo_ataque].tipo in self.debilidades:
                    rival_mult = 2
                elif rival.movimientos[tipo_ataque].tipo in self.fortalezas:
                    rival_mult = .5
                else:
                    rival_mult = 1
                
                print(f"El real que causará según las debilidades y fortalezas es {int(rival_dano * rival_mult)} \n")
                self.vida = self.vida - int(rival_dano * rival_mult)
                
                print(f"{rival.especie} ha hecho {int(rival_dano * rival_mult)} puntos de daño\n")
                print(f"A {self.especie} le quedan {self.vida} puntos de vida\n")
                
                mi_turno = not mi_turno
            round+=1
        
         
        if rival.vida <= 0:
            print(f"¡{rival.especie.title()} se ha desmayado! {self.especie.title()} ha ganado el combate.\n")
        else:
            print(f"¡{self.especie.title()} se ha desmayado! {rival.especie.title()} ha ganado el combate.\n")


```
